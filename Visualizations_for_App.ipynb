# AI Can't Believe It's Not Butter: Fake Data for App Visualizations

```python
import matplotlib.pyplot as plt
from faker import Faker
import random

# Create an instance of Faker
fake = Faker()
```

```python
# Define categories for expenses
expense_categories = ['Groceries', 'Utilities', 'Shopping', 'Eating Out', 'Transportation', 'Entertainment', 'Other']
```

```python
# Generate random monthly expenses
monthly_expenses = {category: random.randint(100, 1000) for category in expense_categories}
```

```python
# Generate total monthly income and calculate total expenses
monthly_income = random.randint(1500, 3000)
total_expenses = sum(monthly_expenses.values())
```

```python
# Pie chart for monthly expenses
plt.figure(figsize=(10, 5))
plt.pie(monthly_expenses.values(), labels=monthly_expenses.keys(), autopct='%1.1f%%', startangle=140)
plt.title('Monthly Expenses')
plt.axis('equal')
plt.show()
```

```python
# Bar chart for income vs. expenses
categories = ['Income', 'Expenses']
amounts = [monthly_income, total_expenses]
```

```python
plt.figure(figsize=(8, 5))
plt.bar(categories, amounts, color=['blue', 'red'])
plt.title('Income vs. Expenses')
plt.ylabel('Amount ($)')
plt.show()
```
